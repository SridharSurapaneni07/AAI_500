## Problem 3.2.

In an exit poll of 1648 voters in the 2020 Senatorial election in Arizona, 51.5% said they voted
for Mark Kelly and 48.5% said they voted for Martha McSally

a) Suppose that actually 50% of the population voted for Kelly. If this exit poll had the
properties of a simple random sample, find the standard error of the sample proportion
voting for him.

b) Under the 50% presumption, are the results of the exit poll surprising? Why? Would you
be willing to predict the election outcome? Explain by (i) conducting a simulation; (ii)
using the value found in (a) for the standard error.

`(a) Your answer goes here`

```python
import math

n = 1648  # Sample size
p = 0.5  # Population proportion voting for Kelly

# Calculate the standard error
standard_error = math.sqrt((p * (1 - p)) / n)

print(f"The standard error of the sample proportion is: {standard_error:.2f}")
```

`(b) Your answer goes here`

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1648
p = 0.5
sample_proportion = 0.515  # Observed sample proportion

# (i) Conducting a simulation
num_simulations = 10000
simulated_proportions = []
for _ in range(num_simulations):
    # Simulate a sample from a binomial distribution with probability p
    sample = np.random.binomial(n, p, 1)
    simulated_proportion = sample / n
    simulated_proportions.append(simulated_proportion[0])

# Plotting the histogram of simulated proportions
plt.hist(simulated_proportions, bins=30)
plt.xlabel("Sample Proportion")
plt.ylabel("Frequency")
plt.title("Distribution of Simulated Sample Proportions")
plt.axvline(sample_proportion, color='red', linestyle='dashed', linewidth=2,
            label=f"Observed Sample Proportion ({sample_proportion:.2f})")
plt.legend()
plt.show()


# (ii) Using the standard error
standard_error = math.sqrt((p * (1 - p)) / n)
z_score = (sample_proportion - p) / standard_error

print(f"The z-score for the observed sample proportion is: {z_score:.2f}")
```

The z-score tells us how many standard deviations away from the mean (0.5) the observed sample proportion is.
A larger z-score indicates a more surprising result.
We can use a z-table to find the probability of observing a sample proportion as extreme or more extreme than the one observed.

# Conclusion:
Based on the simulation, we see that although the observed sample proportion (51.5%) is slightly higher than the assumed population proportion (50%), it falls within a plausible range of outcomes.
Based on the z-score, we can say that the result is not surprising and it is within 1 standard deviation.
We might be willing to predict the election outcome, but we need to be careful because the standard error is still relatively high, and the possibility of error is relatively high.


## Problem 3.3.

 The 49 students in a class at the University of Florida made blinded evaluations of pairs of
cola drinks. For the 49 comparisons of Coke and Pepsi, Coke was preferred 29 times. In the
population that this sample represents, is this strong evidence that a majority prefers Coke?
Use a simulation of a sampling distribution to answer.

`Your answer goes here`

```python
import numpy as np
import matplotlib.pyplot as plt

n = 49  # Sample size
observed_coke_preferences = 29  # Number of times Coke was preferred

# Simulate the sampling distribution under the null hypothesis (50% preference)
num_simulations = 10000
simulated_coke_preferences = []
for _ in range(num_simulations):
    # Simulate a sample from a binomial distribution with probability 0.5
    sample = np.random.binomial(n, 0.5, 1)
    simulated_coke_preferences.append(sample[0])

# Calculate the proportion of simulations where the number of Coke preferences
# is greater than or equal to the observed number of Coke preferences
p_value = np.mean(np.array(simulated_coke_preferences) >= observed_coke_preferences)

# Plotting the histogram of simulated proportions
plt.hist(simulated_coke_preferences, bins=30)
plt.xlabel("Number of Coke Preferences")
plt.ylabel("Frequency")
plt.title("Distribution of Simulated Number of Coke Preferences")
plt.axvline(observed_coke_preferences, color='red', linestyle='dashed', linewidth=2,
            label=f"Observed Number of Coke Preferences ({observed_coke_preferences})")
plt.legend()
plt.show()

print(f"The p-value is: {p_value:.2f}")
```

The p-value represents the probability of observing a sample proportion as extreme or more extreme than the one observed, assuming that the null hypothesis (50% preference) is true.
A smaller p-value indicates stronger evidence against the null hypothesis.

# Conclusion:
Based on the simulation and the p-value, we can conclude that there is some evidence that a majority prefers Coke.
The p-value is not extremely small, but the observed number of Coke preferences is somewhat higher than what would be expected under a 50% preference assumption.
However, we cannot say with high confidence that a majority prefers Coke without further investigation.

## Problem 3.8.

Construct the sampling distribution of the sample proportion of heads, for flipping a balanced
coin (a) once; (b) twice; (c) three times; (d) four times. Describe how the shape changes
as the number of flips $n$ increases. What would happen if $n$ kept growing? Why?


`Your answer goes here`

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

def get_sample_distribution(n_flips):
    all_outcomes = []
    for i in range(2**n_flips):
        binary_representation = bin(i)[2:].zfill(n_flips)
        heads_count = binary_representation.count('1')
        sample_proportion =  round(heads_count / n_flips, 2)
        all_outcomes.append(sample_proportion)
    return Counter(all_outcomes)

# (a) Once
sample_distribution_1 = get_sample_distribution(1)
print("Sampling distribution for 1 flip:", sample_distribution_1)
plt.bar(sample_distribution_1.keys(), sample_distribution_1.values())
plt.title("Sampling Distribution for 1 Flip")
plt.xlabel("Sample Proportion of Heads")
plt.ylabel("Frequency")
plt.show()

# (b) Twice
sample_distribution_2 = get_sample_distribution(2)
print("Sampling distribution for 2 flips:", sample_distribution_2)
plt.bar(sample_distribution_2.keys(), sample_distribution_2.values())
plt.title("Sampling Distribution for 2 Flips")
plt.xlabel("Sample Proportion of Heads")
plt.ylabel("Frequency")
plt.show()

# (c) Three times
sample_distribution_3 = get_sample_distribution(3)
print("Sampling distribution for 3 flips:", sample_distribution_3)
plt.bar(sample_distribution_3.keys(), sample_distribution_3.values())
plt.title("Sampling Distribution for 3 Flips")
plt.xlabel("Sample Proportion of Heads")
plt.ylabel("Frequency")
plt.show()

# (d) Four times
sample_distribution_4 = get_sample_distribution(4)
print("Sampling distribution for 4 flips:", sample_distribution_4)
plt.bar(sample_distribution_4.keys(), sample_distribution_4.values())
plt.title("Sampling Distribution for 4 Flips")
plt.xlabel("Sample Proportion of Heads")
plt.ylabel("Frequency")
plt.show()
```

# Description of shape changes:
As n increases, the sampling distribution becomes more symmetric and bell-shaped.
This is due to the Central Limit Theorem, which states that the sampling distribution of
the sample mean (or proportion) approaches a normal distribution as the sample size increases.

# What would happen if n kept growing?
If n kept growing, the sampling distribution would become even more closely approximated by a normal distribution with a mean of 0.5 (since the coin is balanced) and a standard deviation that decreases as n increases. This means that the sample proportions would be clustered more tightly around 0.5 as the number of flips increases.

# Why?
This happens because with a larger sample size, the effects of random fluctuations in individual flips become less significant, and the overall proportion of heads tends to converge towards the true population proportion (0.5 in this case). The Central Limit Theorem mathematically explains this phenomenon.

## Problem 3.13.

Simulate random sampling from a uniform population distribution with several $n$ values to
illustrate the Central Limit Theorem.

`Your answer goes here`

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_sampling_distribution(population_size, sample_size, num_simulations):
    population = np.random.uniform(0, 1, population_size)  # Uniform population
    sample_means = []

    for _ in range(num_simulations):
        sample = np.random.choice(population, sample_size, replace=False)
        sample_means.append(np.mean(sample))

    return sample_means

# Parameters for the simulation
population_size = 10000
num_simulations = 10000
sample_sizes = [5, 10, 25, 50]

# Simulate and plot the sampling distribution for different sample sizes
for sample_size in sample_sizes:
    sample_means = simulate_sampling_distribution(population_size, sample_size, num_simulations)

    plt.hist(sample_means, bins=30, density=True, alpha=0.7, label=f'n = {sample_size}')

plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Sampling Distribution of the Sample Mean (Uniform Population)')
plt.legend()
plt.show()

```

`Your answer goes here`

The approximate distribution of Y is a normal distribution with mean 0 and standard deviation sqrt(100).

# Reasoning:
Each bet is an independent event with a probability of `0.5 of winning $1 and 0.5 of losing $1.`
The expected value of each bet is `0.5 * $1 + 0.5 * (-$1) = $0.`
The variance of each bet is `0.5 * ($1 - $0)^2 + 0.5 * (-$1 - $0)^2 = $1.`
The total winnings and losings after 100 bets is the sum of 100 independent bets.
By the Central Limit Theorem, the sum of a large number of independent random variables with finite mean and variance tends towards a normal distribution.
Therefore, the approximate distribution of Y is a normal distribution with mean `100 * $0 = $0 and variance 100 * $1 = $100.`
`The standard deviation of Y is sqrt(100) = $10.`

In conclusion, the approximate distribution of Y is a normal distribution with mean` $0 and standard deviation $10.`

## Problem 3.21.

In your school, suppose that GPA has an approximate normal distribution with $\mu = 3.0, \sigma = 0.40.$ Not knowing $\mu$, you randomly sample $n = 25$ students to estimate it. Using simulation for this application, illustrate the difference between a sample data distribution and the sampling distribution of Y .

`Your answer here`

```python
import numpy as np
import matplotlib.pyplot as plt

# Population parameters
mu = 3.0
sigma = 0.40
n = 25

# Simulate a sample data distribution
sample_data = np.random.normal(loc=mu, scale=sigma, size=n)

# Plot the sample data distribution
plt.hist(sample_data, bins=10, density=True, alpha=0.7, label='Sample Data Distribution')
plt.xlabel('GPA')
plt.ylabel('Density')
plt.title('Sample Data Distribution (n=25)')
plt.legend()
plt.show()

# Simulate the sampling distribution of the sample mean
num_simulations = 10000
sample_means = []
for _ in range(num_simulations):
    sample = np.random.normal(loc=mu, scale=sigma, size=n)
    sample_means.append(np.mean(sample))

# Plot the sampling distribution of the sample mean
plt.hist(sample_means, bins=20, density=True, alpha=0.7, label='Sampling Distribution of Ȳ')
plt.xlabel('Sample Mean (Ȳ)')
plt.ylabel('Density')
plt.title('Sampling Distribution of the Sample Mean (n=25)')
plt.legend()
plt.show()

# Print some descriptive statistics
print(f"Sample Data Mean: {np.mean(sample_data):.2f}")
print(f"Sample Data Standard Deviation: {np.std(sample_data):.2f}")
print(f"Sampling Distribution Mean: {np.mean(sample_means):.2f}")
print(f"Sampling Distribution Standard Deviation (Standard Error): {np.std(sample_means):.2f}")
```
